```python
import glob as glob
```

```python
# importing NYC, BOSTON and AUSTIN Airbnb lisings
city_listings = []
```

```python
#cities = glob.glob("[a-z][a-z][a-z]_listings.csv")
cities = glob.glob("[a-z]_listings.csv")
```

```python
cities
```

```python
for city in cities:
    all_cities = pd.read_csv(city)
    city_listings.append(all_cities)
    
airbnb = pd.concat(city_listings)
```

```python
airbnb.head()
```

```python
airbnb.info()
```

```python
airbnb.isnull().sum().sort_values(ascending=False)
```

```python
# Boston and Austin do not have neighbourhood groups.
```

```python
# replacing null values in reviews per month to 0. This is a probably the rating
# replacing host name with unknown
airbnb.fillna({'reviews_per_month':0, 'host_name': 'unknown'}, inplace=True)
```

```python
# dropping last review column as it is the date of last review.
airbnb.drop(['last_review'], axis = 1,inplace = True)
```

```python
# Dropping listing name
airbnb.drop(['name','Unnamed: 0'], axis = 1,inplace = True)
```
